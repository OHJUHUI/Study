## 1. 데이터 수집

1) 데이터 품질

2) 데이터 형식

(1) CSV(Comma Separated Values)

- CSV는 각 항목 값을 콤마 기호(,)로 구분하는 데이터

- 가장 효율적으로 데이터를 텍스트 파일로 저장할 수 있음
→ 때문에 분석용 데이터 포맷 중 가장 많이 사용됨

- 마이크로소프트 엑셀과 같은 도구로 쉽게 변환하여 추출할 수 있음

(2) XLSX

- 마이크로소프트 엑셀에서 만든 XML(eXtensible Markup Language)기반 파일 포맷

- 압축된 형태로 저장 → 일반 텍스트 에디터로는 열람할 수 없음
→엑셀,애플 넘버스, 구글 문서 등으로 열람할 수 있음

- 개발 지식이 없는 일반 사용자에게 가장 적합한 포맷

(3) JSON(JavaScript Object Notation)

- 자바스크립트에서 사용하는 데이터 표현 방식을 그대로 사용한 포맷

 → 자바스크립트에서만 사용할 수 있는 것은 아님

 → 인터넷 네트워크로 데이터를 전송할 때 사실상 표준으로 자리 잡은 데이터 전송 포맷

- 파일크기는 CSV나 XLSX보다 훨씬 크지만 값을 표현하는 방식이 굉장히 유연함

 → 이러한 특징으로 많은 곳에서 표준 포맷으로 사용함

### 확진자 비율 구하기

• 인구 대비 확진자 비율 구하기

 → 단순 확진자 수로 데이터를 비교하면 의미가 왜곡될 수 있음

 → 각국 인구 데이터 비교 → 데이터프레임 population 행의 2022년 1월 1일자 인구를 확인

```python
# 대한민국 인구
kor_population = kor_date_index_df['population']['2022-01-22']
# 미국 인구
usa_population = usa_date_index_df['population']['2022-01-22']

print('미국 인구(', usa_population, ') / 대한민국 인구(', kor_population, ')')
```

• 인구 대비 확진자 비율 구하기

→ 미국 인구를 대학민국 인구로 나눠 비율 확인

→ 소수점 2자리 까지 표현→ 파이썬 기본 함수 round()로 반올림

```python
# 인구 비율 구하기
rate = round((usa_population / kor_population), 2)
rate
```

### 확진자 비율 반영한 그래프 그리기

• 확진자 비율 반영한 데이터프레임 생성하기
