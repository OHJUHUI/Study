```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
pig = pd.read_excel('/content/(첨부1)(정제)충청남도_멧돼지포획실적_200409~230331_호서대.xlsx')
pig
```

```python
pig_df = ['위도', '경도', '포획일', '시군']
pig_columns = pig[pig_df]
pig_columns
```

```python
pig_year = pig_columns.groupby([pig_columns['포획일'].dt.year]).size()
pig_year
```

```python
pig_year.plot.barh()
plt.title('연도별 포획수 비교', fontsize=15)
plt.xlabel('포획수', fontsize=10)
plt.ylabel('연도', fontsize=10)
plt.show()
```

```python
pig_month = pig_columns.groupby([pig_columns['포획일'].dt.month]).size()
pig_month
```

```python
pig_month.plot.barh()
plt.title('월별 포획수 비교', fontsize=15)
plt.xlabel('포획수', fontsize=10)
plt.ylabel('월', fontsize=10)
plt.show()
```
