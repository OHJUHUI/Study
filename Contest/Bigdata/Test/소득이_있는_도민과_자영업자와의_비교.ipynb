```python
# 'CRTR_YM' 열 데이터를 해석하여 'YEAR' 열에 연도를 저장하는 함수
def extract_year(crtr_ym_str):
    if crtr_ym_str.startswith('2016'):
        return 2016
    elif crtr_ym_str.startswith('2017'):
        return 2017
    elif crtr_ym_str.startswith('2018'):
        return 2018
    elif crtr_ym_str.startswith('2019'):
        return 2019
    elif crtr_ym_str.startswith('2020'):
        return 2020
    elif crtr_ym_str.startswith('2021'):
        return 2021
    elif crtr_ym_str.startswith('2022'):
        return 2022
    elif crtr_ym_str.startswith('2023'):
        return 2023
    else:
        return None

# 'CRTR_YM' 열 데이터를 해석하여 'Year' 열에 연도를 저장
info_data['YEAR'] = info_data['CRTR_YM_STR'].apply(extract_year)
```
