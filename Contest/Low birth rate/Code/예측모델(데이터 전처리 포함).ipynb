- **경제활동 인구 데이터 전처리(coin)**

```python
coin = coin[['시점', '계.1', '계.2', '계.3', '계.4', '계.5', '계.6', '계.7']]
coin
```

```python
coin.columns = coin.iloc[0]
coin = coin[1:].reset_index(drop=True)
coin
```

```python
coin = coin.rename(columns={'경제활동인구 (천명)':'경제활동인구',	'취업자 (천명)':'취업자',	'실업자 (천명)': '실업자',	'비경제활동인구 (천명)':'비경제활동인구',	'경제활동참가율 (%)':'경제활동참가율',	'실업률 (%)':'실업률', '고용률 (%)':'고용률'})
coin = coin.drop([0], axis=0)
coin
```

- 경제활동인구, 취업자, 실업자, 비경제 활동인구, 경제활동참가율, 실업률, 고용률 데이터

- **all 데이터 전처리**

```python
all.columns = all.iloc[0]
all = all[1:].reset_index(drop=True)
```

```python
all
```

```python
unique_values = all['종류별'].unique()
unique_values
```

```python
all['시점'] = pd.to_datetime(all['시점'].str.extract('(\d+\.\d+)')[0], format='%Y.%m', errors='coerce')
all
```

```python
# '행정구역별'이 '전국'인 데이터만 추출
filtered_all = all[all['행정구역별'] == '전국']
filtered_all
```

- 출생아 수

```python
baby = filtered_all[filtered_all['종류별'] == '출생아수(명)']
baby = baby.drop(columns=['종류별', '행정구역별'])
baby = baby.rename(columns={'출생,사망,혼인,이혼': '출생아수'})
baby
```

- 사망자 수

```python
# 사망자수 데이터값 1981-1982년 사망자 데이터 없음 확인완료!!

die = filtered_all[filtered_all['종류별'] == '사망자수(명)']
die = die.drop(columns=['종류별', '행정구역별'])
die = die.rename(columns={'출생,사망,혼인,이혼': '사망자수'})
die
```

- 혼인 수

```python
marry = filtered_all[filtered_all['종류별'] == '혼인건수(건)']
marry = marry.drop(columns=['종류별', '행정구역별'])
marry = marry.rename(columns={'출생,사망,혼인,이혼': '혼인 수'})
marry
```

- 이혼 수

```python
divorce = filtered_all[filtered_all['종류별'] == '이혼건수(건)']
divorce = divorce.drop(columns=['종류별', '행정구역별'])
divorce = divorce.rename(columns={'출생,사망,혼인,이혼': '이혼 수'})
divorce
```
