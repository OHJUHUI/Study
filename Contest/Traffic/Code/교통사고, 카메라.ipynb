```python
import folium
from folium.features import Marker

def marker_color(row):
    if row['day_night'] == 'night':
        return 'red'
    elif row['day_night'] == 'late evening':
        return 'blue'
    elif row['day_night'] == 'early evening':
        return 'green'
    elif row['day_night'] == 'early morning':
        return 'orange'
    elif row['day_night'] == 'late morning':
        return 'black'
    elif row ['day_night']=='early afternoon':
        return 'purple'
    elif row ['day_night']== 'late afternoon':
        return 'brown'

kp_df['marker_color'] = kp_df.apply(marker_color, axis=1)

m = folium.Map(location=[kp_df['HPPN_Y'].mean(), kp_df['HPPN_X'].mean()], zoom_start=10)

for i, row in kp_df.iterrows():
    folium.CircleMarker(
        location=[row['HPPN_Y'], row['HPPN_X']],
        radius=5,
        color=row['marker_color'],
        fill=True,
        fill_color=row['marker_color']
    ).add_to(m)

m
```

```python
m.save('accident_map.html')
```
